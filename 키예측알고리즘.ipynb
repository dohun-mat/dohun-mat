```python
import tensorflow as tf
```

```python
키 = 170
신발 = 260

#신발 = 키 * a + b
a = tf.Variable(0.1)
b = tf.Variable(0.2)

def 손실함수():
    #실제값 - 예측값
    return tf.square(신발 - (키 * a + b))

opt = tf.keras.optimizers.Adam(learning_rate=0.1)
#경사하강법도 공부필요

for i in range(300):
    print("a : \n",a.numpy())
    print("b :",b.numpy())
    opt.minimize(손실함수, var_list=[a,b])
    


```
